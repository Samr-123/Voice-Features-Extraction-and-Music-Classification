```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from PIL import Image

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.DataFrame(columns=['mfcc','rms','chroma_stft','spectral_centroid','tonnetz','chroma_cens','spectral_bandwidth','spectral_flatness'])
for i in range(1,21):
    df[f'mfcc{i}']=""
df['genre']=""
```

```python
df
```

```python
genres=['classical','pop']
```

```python
import os
```

```python
all_features=[]
for genre in genres:
    for filename in os.listdir(f'mini-genres/{genre}'):
        songname=f'mini-genres/{genre}/{filename}'
        y,sr=librosa.load(songname,mono=True,duration=30)
        feature_mfccs=librosa.feature.mfcc(y,sr=sr)
        feature_rms=librosa.feature.rms(y=y)
        feature_chroma=librosa.feature.chroma_stft(y,sr=sr)
        centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
        tonnetz = librosa.feature.tonnetz(y=y, sr=sr)
        chroma_cens = librosa.feature.chroma_cens(y=y, sr=sr)
        spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
        flatness = librosa.feature.spectral_flatness(y=y)
        feature_list=[np.mean(feature_mfccs),np.mean(feature_rms),np.mean(feature_chroma),np.mean(centroid),np.mean(tonnetz),
                      np.mean(chroma_cens),np.mean(spec_bw),np.mean(flatness)]
        for feature in feature_mfccs:
            feature_list.append(np.mean(feature))
        feature_list.append(genre)
        all_features.append(feature_list)
        
```

```python
df=pd.DataFrame(all_features,columns=df.columns)
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(np.array(df.iloc[:,:-1],dtype=float))
```

```python
label_encoder=LabelEncoder()
```

```python
Y=label_encoder.fit_transform(df['genre'])
```

```python
X_train,X_test, Y_train,Y_test=train_test_split(X,Y, test_size=0.3, random_state=346)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score 
```

```python
print(accuracy_score(Y_test,pred)*100)
```
